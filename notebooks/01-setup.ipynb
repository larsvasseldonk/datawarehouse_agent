```python
import os
import urllib.request

import duckdb

DATA_URL = "https://blobs.duckdb.org/nl-railway/services-2025-03.csv.gz"
CSV_FILE = "services-2025-03.csv.gz"

con = duckdb.connect("dutch_railways.db")


def setup_database():
    """Download the CSV file and load it into DuckDB."""
    # if not os.path.exists(CSV_FILE):
    #     print(f"Downloading {DATA_URL}...")
    #     urllib.request.urlretrieve(DATA_URL, CSV_FILE)

    con.execute(f"""
        CREATE TABLE IF NOT EXISTS services AS
        SELECT * FROM '{DATA_URL}'
    """)
    count = con.execute("SELECT COUNT(*) FROM services").fetchone()[0]
    print(f"Loaded {count} rows")
    return count
```

```python
setup_database()
```

```python
con.execute("DESCRIBE services").fetchall()
```
